Converting JSON format data to individual columns

```python
import csv
import ast
import pandas as pd

with open('DataAnalyst-Jobs-USA.csv', 'r') as file:
    reader = csv.reader(file)
    header_row = next(reader)
    column_names = ['Seniority level', 'Employment type', 'Job function', 'Industries']
    header_row[6:6] = column_names
    with open('output.csv', 'w', newline='') as outfile:
        writer = csv.writer(outfile)
        writer.writerow(header_row)
    for row in reader:
        if len(row) >= 7:
            dictionary_list_string = row[6]
            # This function converts the string into dictionary,
            dictionary_list = ast.literal_eval(dictionary_list_string)
            dictionary = {}
            for dictionary_item in dictionary_list:
                for key, value in dictionary_item.items():
                    dictionary[key] = value
            values = [dictionary.get(key, '') for key in column_names]
            row[6:6] = values
            with open('output.csv', 'a', newline='') as outfile:
                writer = csv.writer(outfile)
                writer.writerow(row)
        else:
            with open('output.csv', 'a', newline='') as outfile:
                writer = csv.writer(outfile)
                writer.writerow(row)

df=pd.read_csv('output.csv')
df.dropna(inplace=True)
df
```

Transforming the salary column

```python
# loop through each row in the salary column
for i, row in df.iterrows():
    # check if '-' is present in the string
    if '-' in row['salary']:
        # if yes, split the string into two columns
        salary_min, salary_max = row['salary'].split('-')
        # remove '/yr' and '$' from the columns
        df.at[i, 'salary_min'] = salary_min.replace('/yr', '').replace('$', '').replace(',', '').strip()
        df.at[i, 'salary_max'] = salary_max.replace('/yr', '').replace('$', '').replace(',', '').strip()
        if pd.to_numeric(df.at[i, 'salary_min']) < 200:
          df.at[i, 'salary_type'] = "hourly"
        else:
          df.at[i, 'salary_type'] = "annual"
    else:
        # if not, set both columns to the same value
        salary = row['salary'].replace('/yr', '').replace('$', '').replace(',', '').strip()
        df.at[i, 'salary_min'] = salary
        df.at[i, 'salary_max'] = salary
        if pd.to_numeric(df.at[i, 'salary_min']) < 200:
          df.at[i, 'salary_type'] = "hourly"
        else:
          df.at[i, 'salary_type'] = "annual"

# convert both columns to float
df['salary_min'] = pd.to_numeric(df['salary_min'])
df['salary_max'] = pd.to_numeric(df['salary_max'])

df.loc[df['salary'].str.contains('/yr', case=False), 'salary_type'] = 'annual'
df.loc[df['salary'].str.contains('/hr', case=False), 'salary_type'] = 'hourly'
# drop the original salary column
df.drop('salary', axis=1, inplace=True)

# print the updated DataFrame
df
```

Transforming the location column

```python
import pandas as pd

# Create empty columns for city, state, and country
df["city"] = ""
df["state"] = ""
df["country"] = "United States"
# Loop through the rows of the DataFrame
for i, row in df.iterrows():
    # Split the location string into city and state/country
    location = row["location"]
    if "," not in location:
      if location not in state_full_form and location != "United States":
        df.at[i,"city"] = location
        df.at[i,"state"] = "Unknown"
      elif location in state_full_form:
        df.at[i,"state"] = location
        df.at[i,"city"] = "Unknown"
    if "," in location:
        city, state = location.split(", ")
        # Modify the city and state/country columns based on the scenario
        if city not in state_full_form:
            df.at[i, "city"] = city.strip()
            if state in state_dict:
              df.at[i, "state"] = state_dict[state]
            if state in state_full_form:
              df.at[i, "state"] = state
        else:
          df.at[i, "city"] = "Unknown"
          if city in state_dict:
              df.at[i, "state"] = state_dict[city]
          else:
              df.at[i, "state"] = city

df["state"] = df["state"].fillna("Unknown")
# Print the modified DataFrame

df=df.drop("link",axis=1)
df=df.drop("criteria",axis=1)
df=df.drop("location",axis=1)
df=df.drop("description",axis=1)
df=df.drop("Industries",axis=1)

df = df.rename(columns={'Seniority level': 'seniority_level', 'Employment type': 'employment_type', 'Job function': 'job_function'})
df
```

```python
import pandas as pd

job = pd.read_csv("LinkedIn-Jobs.csv")
job = job.drop("context",axis=1)
job = job.drop("description",axis=1)
job = job.drop("post_url",axis=1)
job = job.drop("post_id",axis=1)
job = job.drop("company_id",axis=1)

job["city"] = ""
job["state"] = ""
job["country"] = "United States"

for i, row in job.iterrows():
    # Split the location string into city and state/country
    location = row["location"]
    if "," not in location:
      if location not in state_full_form and location != "United States":
        job.at[i,"city"] = location
        job.at[i,"state"] = "Unknown"
      elif location in state_full_form:
        job.at[i,"state"] = location
        job.at[i,"city"] = "Unknown"
    if "," in location:
        city, state = location.rsplit(",", 1)
        city=city.strip()
        state=state.strip()
        # Modify the city and state/country columns based on the scenario
        if city not in state_full_form:
            job.at[i, "city"] = city.strip()
            if state in state_dict:
              job.at[i, "state"] = state_dict[state]
            if state in state_full_form:
              job.at[i, "state"] = state
        else:
          job.at[i, "city"] = "Unknown"
          if city in state_dict:
              job.at[i, "state"] = state_dict[city]
          else:
              job.at[i, "state"] = city

job=job.drop("location",axis=1)
job
```

```python
import numpy as np

job['salary'].fillna(0, inplace=True)

# filter the rows that contain only "$" and remove any conversion strings
job = job[job['salary'].str.contains('\$', na=False)].copy()
job['salary'] = job['salary'].str.replace(r'\([^()]*\)', '') # remove any currency conversion strings in parentheses

# split salary range into salary_min and salary_max
job[['salary_min', 'salary_max']] = job['salary'].str.split(' - ', expand=True)

# remove the "$" and "/yr" strings from the salary columns and convert to numeric type
job['salary_min'] = pd.to_numeric(job['salary_min'].str.replace('$', '').str.replace(',', '').str.replace('/yr', '').str.replace('/hr', ''))
job['salary_max'] = pd.to_numeric(job['salary_max'].str.replace('$', '').str.replace(',', '').str.replace('/yr', '').str.replace('/hr', ''))

# add a column indicating whether the salary is in annual pay or hourly pay
job['salary_type'] = np.where(job['salary'].str.contains('/hr'), 'hourly', 'annual')

job = job.drop("salary",axis=1)
job
```

```python
import nltk
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
from nltk import word_tokenize, pos_tag

def generalize_degree(degree):
    tokens = word_tokenize(degree.lower())
    
    tagged_tokens = pos_tag(tokens)
    
    generalized_terms = []
    
    for i in range(len(tagged_tokens)):
        if tagged_tokens[i][1] == 'NN' or tagged_tokens[i][1] == 'NNS':
            if tagged_tokens[i][0] in ['bachelor', 'bs', 'b.e.', 'bachelore', 'bachelors', 'btech', 'gniit', 
                                       'master', 'ms', 'm.s.', 'm.e.', 'm.sc.', 'm.t.', 'mba', 'pgdm', 'mca',
                                       'm.eng.', 'm.tech', 'phd', 'jd', 'grduate', 'executive', 'fellowship', 
                                       'global', 'product', 'doctorate']:
                if tagged_tokens[i][0] in ['bachelor', 'bs', 'b.e.', 'bachelore', 'bachelors', 'bachelor\'s']:
                    generalized_terms.append('Bachelor\'s')
                elif tagged_tokens[i][0] in ['master', 'masters', 'ms', 'm.s.']:
                    generalized_terms.append('MS')
                elif tagged_tokens[i][0] in ['grduate']:
                    generalized_terms.append('Graduate')
                else:
                    generalized_terms.append(tagged_tokens[i][0])
    
    generalized_terms = list(set(generalized_terms))
    
    generalized_string = ' '.join(generalized_terms)
    
    return generalized_string

pf['Degree'] = pf['Degree'].apply(generalize_degree)
print(pf['Degree'].unique())

pf[['Degree_1', 'Degree_2']] = pf['Degree'].str.split(' ', n=1, expand=True)

pf['Degree_2'].fillna('Unknown', inplace=True)

pf = pf.drop('Degree', axis=1)
pf = pf.drop('Degree of Connection', axis=1)
```

```python
month_names = ['January', 'February', 'March', 'April', 'May', 'June','July','August', 'September', 'October', 'November', 'December']
pf['Month Started'] = pf['Month Started'].astype(int).astype(str)
pf['Month Started'] = pf['Month Started'].apply(lambda x: month_names[int(x)-1])

pf = pf.rename(columns={'First Name': 'first_name', 'Last Name': 'last_name', 'Full Name': 'full_name' , 'Job Title': 'job_title', 'Company Name': 'company_name', 'New Job (90 Days)': 'new_job', 'Current Position': 'current_position', 'Month Started': 'month_started','Year Started': 'year_started', 'School': 'school', 'Education Start': 'education_start','Education End': 'education_end' , 'Field of Study': 'field', 'Shared Connections': 'shared_connections', 'Domain': 'domain' , 'Degree_1': 'degree_1', 'Degree_2': 'degree_2' })
pf.to_csv("profile.csv")
```
